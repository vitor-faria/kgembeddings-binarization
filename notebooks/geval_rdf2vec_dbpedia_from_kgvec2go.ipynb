## Get DBpedia entities from GEval datasets

```python
import pandas as pd


datasets = [
    # Classification
    'evaluation_framework/Classification/data/AAUP.tsv',
    'evaluation_framework/Classification/data/Cities.tsv',
    'evaluation_framework/Classification/data/Forbes.tsv',
    'evaluation_framework/Classification/data/MetacriticAlbums.tsv',
    'evaluation_framework/Classification/data/MetacriticMovies.tsv',
    # Clustering
    'evaluation_framework/Clustering/data/cities2000AndCountries_cluster.tsv',
    'evaluation_framework/Clustering/data/citiesAndCountries_cluster.tsv',
    'evaluation_framework/Clustering/data/citiesMoviesAlbumsCompaniesUni_cluster.tsv',
    'evaluation_framework/Clustering/data/teams_cluster.tsv',
    # Classification
    'evaluation_framework/Regression/data/AAUP.tsv',
    'evaluation_framework/Regression/data/Cities.tsv',
    'evaluation_framework/Regression/data/Forbes.tsv',
    'evaluation_framework/Regression/data/MetacriticAlbums.tsv',
    'evaluation_framework/Regression/data/MetacriticMovies.tsv',
]

entity_set = set()

for dataset in datasets:
    df = pd.read_csv(dataset, sep='\t')
    dataset_entities = set(df['DBpedia_URI'].unique())
    print(f'{len(dataset_entities)} unique entities in dataset {dataset}')
    entity_set.update(dataset_entities)
    print(f'{len(entity_set)} unique entities in total.')
```

```python
semantic_datasets = [
    'evaluation_framework/SemanticAnalogies/data/all_capital_country_entities.txt',
    'evaluation_framework/SemanticAnalogies/data/capital_country_entities.txt',
    'evaluation_framework/SemanticAnalogies/data/city_state_entities.txt',
    'evaluation_framework/SemanticAnalogies/data/currency_entities.txt',
]

semantic_entity_set = set()
for dataset in semantic_datasets:
    dataset_entities = set()
    df = pd.read_csv(
        dataset,
        sep=' ',
        names=['entity_1', 'entity_2', 'entity_3', 'entity_4'],
    )
    dataset_entities.update(set(df['entity_1'].unique()))
    dataset_entities.update(set(df['entity_2'].unique()))
    dataset_entities.update(set(df['entity_3'].unique()))
    dataset_entities.update(set(df['entity_4'].unique()))
    print(f'{len(dataset_entities)} unique entities in dataset {dataset}')
    semantic_entity_set.update(dataset_entities)
    print(f'{len(semantic_entity_set)} unique entities in total.')
```

```python
import json

with open('evaluation_framework/DocumentSimilarity/data/LP50_entities.json', 'r') as file:
    doc_sim_dataset = json.load(file)

doc_sim_entity_set = set()
for text in doc_sim_dataset:
    for annotation in text['annotations']:
        doc_sim_entity_set.add(annotation['entity'])

len(doc_sim_entity_set)
```

```python
ent_rel_df = pd.read_csv(
    'evaluation_framework/EntityRelatedness/data/KORE.txt',
    sep='\t',
    names=['index', 'DBpedia_URI']
)

ent_rel_set = set()
ent_rel_set.update(set([
    ent for ent in ent_rel_df['index'].tolist() if isinstance(ent, str)
]))
ent_rel_set.update(set([
    ent for ent in ent_rel_df['DBpedia_URI'].tolist() if isinstance(ent, str)
]))

len(ent_rel_set)
```

```python
entity_set.update(semantic_entity_set)
entity_set.update(doc_sim_entity_set)
entity_set.update(ent_rel_set)
len(entity_set)
```

## Get binarized embeddings from autoencoder

```python
!git clone https://github.com/vitor-faria/kgembeddings-binarization.git binarizer
```

```python
vec_sizes = [128, 256, 512]

def int_to_bitlist(int_str):
    bitlist = [bit for bit in bin(int(int_str))[2:]]
    missing_zeros = 64 - len(bitlist)
    return ['0'] * missing_zeros + bitlist

def get_full_binary_vector(line):
    bin_vector = []
    for int_str in line.split()[1:]:
        bin_vector.extend(int_to_bitlist(int_str))
    return line.split()[0] + ' ' + ' '.join(bin_vector)

for vec_size in vec_sizes:

    with open(f"binarizer/binary_vectors/binary_vectors_rdf2vec_{vec_size}.vec", "r") as file:
        lines = file.readlines()

    with open(f"autoencoded_rdf2vec_bin_{vec_size}.txt", "w") as file:
        for line in lines[1:]:
            full_binary_vector = get_full_binary_vector(line)
            file.write(f'{full_binary_vector} \n')
```
