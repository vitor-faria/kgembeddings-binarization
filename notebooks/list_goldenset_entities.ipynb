```python
import pandas as pd


datasets = [
    # Classification
    'geval/evaluation_framework/Classification/data/AAUP.tsv',
    'geval/evaluation_framework/Classification/data/Cities.tsv',
    'geval/evaluation_framework/Classification/data/Forbes.tsv',
    'geval/evaluation_framework/Classification/data/MetacriticAlbums.tsv',
    'geval/evaluation_framework/Classification/data/MetacriticMovies.tsv',
    # Clustering
    'geval/evaluation_framework/Clustering/data/cities2000AndCountries_cluster.tsv',
    'geval/evaluation_framework/Clustering/data/citiesAndCountries_cluster.tsv',
    'geval/evaluation_framework/Clustering/data/citiesMoviesAlbumsCompaniesUni_cluster.tsv',
    'geval/evaluation_framework/Clustering/data/teams_cluster.tsv',
    # Regression
    'geval/evaluation_framework/Regression/data/AAUP.tsv',
    'geval/evaluation_framework/Regression/data/Cities.tsv',
    'geval/evaluation_framework/Regression/data/Forbes.tsv',
    'geval/evaluation_framework/Regression/data/MetacriticAlbums.tsv',
    'geval/evaluation_framework/Regression/data/MetacriticMovies.tsv',
]

geval_entity_set = set()

for dataset in datasets:
    df = pd.read_csv(dataset, sep='\t')
    dataset_entities = set(df['DBpedia_URI'].unique())
    print(f'{len(dataset_entities)} unique entities in dataset {dataset}')
    geval_entity_set.update(dataset_entities)
    print(f'{len(geval_entity_set)} unique entities in total.')
```

```python
semantic_datasets = [
    'geval/evaluation_framework/SemanticAnalogies/data/all_capital_country_entities.txt',
    'geval/evaluation_framework/SemanticAnalogies/data/capital_country_entities.txt',
    'geval/evaluation_framework/SemanticAnalogies/data/city_state_entities.txt',
    'geval/evaluation_framework/SemanticAnalogies/data/currency_entities.txt',
]

semantic_entity_set = set()
for dataset in semantic_datasets:
    dataset_entities = set()
    df = pd.read_csv(
        dataset,
        sep=' ',
        names=['entity_1', 'entity_2', 'entity_3', 'entity_4'],
    )
    dataset_entities.update(set(df['entity_1'].unique()))
    dataset_entities.update(set(df['entity_2'].unique()))
    dataset_entities.update(set(df['entity_3'].unique()))
    dataset_entities.update(set(df['entity_4'].unique()))
    print(f'{len(dataset_entities)} unique entities in dataset {dataset}')
    semantic_entity_set.update(dataset_entities)
    print(f'{len(semantic_entity_set)} unique entities in total.')
```

```python
import json

with open('geval/evaluation_framework/DocumentSimilarity/data/LP50_entities.json', 'r') as file:
    doc_sim_dataset = json.load(file)

doc_sim_entity_set = set()
for text in doc_sim_dataset:
    for annotation in text['annotations']:
        doc_sim_entity_set.add(annotation['entity'])

len(doc_sim_entity_set)
```

```python
ent_rel_df = pd.read_csv(
    'geval/evaluation_framework/EntityRelatedness/data/KORE.txt',
    sep='\t',
    names=['index', 'DBpedia_URI']
)

ent_rel_set = set()
ent_rel_set.update(set([
    ent for ent in ent_rel_df['index'].tolist() if isinstance(ent, str)
]))
ent_rel_set.update(set([
    ent for ent in ent_rel_df['DBpedia_URI'].tolist() if isinstance(ent, str)
]))

len(ent_rel_set)
```

```python
geval_entity_set.update(semantic_entity_set)
geval_entity_set.update(doc_sim_entity_set)
geval_entity_set.update(ent_rel_set)
len(geval_entity_set)
```

```python
import pickle

with open('geval_entities.pickle', 'wb') as fp:
    pickle.dump(geval_entity_set, fp)
```

## List DLCC entities

```python
!git clone https://github.com/janothan/DL-TC-Generator.git dlcc
```

```python
import os

dlcc_entities = set()
directory = 'dlcc/results/v1/dbpedia'
for dirpath, dirnames, filenames in os.walk(directory):
    for filename in filenames:
        if filename.endswith('.txt'):
            print(os.path.join(dirpath, filename))
            with open(os.path.join(dirpath, filename)) as f:
                file_entities = set(f.read().split())
                dlcc_entities.update(file_entities)
                print(len(dlcc_entities))
```

```python
import pickle

with open('dlcc_entities.pickle', 'wb') as fp:
    pickle.dump(dlcc_entities, fp)
```
